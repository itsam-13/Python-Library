```python
import pandas as pd
import numpy as np
```

<h4>Finding Missing Data</h4>

```python
data = {
'A': [1, 2, np.nan, 4, 5],
'B': [1, 2, 3, 4, 5],
'C': [1, 2, 3, np.nan, np.nan],
'D': [1, np.nan, np.nan, np.nan, 5]
}

df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum() #count of missing values in each column
```

```python
df.isna().any()    #check if any missing values in each column
```

<i>Removing Missing Data</i>

```python
df
```

```python
df.dropna()  #drop rows with any missing values. dropna-> row wise
```

```python
df.dropna(thresh=3)  #drop rows with at least 3 non-missing values
```

<i>Filling the Missing Data</i>

```python
df
```

```python
df.fillna(0)  #fill missing values with 0. fillna-> target column wise
```

```python
values = {'A': 0, 'B': 100, "C": 300, 'D': 400}
df.fillna(value=values)  #fill missing values with different values for each column
```

```python
df
```

```python
df.fillna(df.mean())
```
