```python
import numpy as np
```

NUMPY INDEXING AND SLICING OF VECTORS

```python
arr = np.arange(1,30,3)  # start, stop, step
arr
```

```python
arr[4]  # access element at index 4
```

```python
arr[1:5] # slicing from index 1 to 4
```

```python
arr[::2]  # slicing with step size 2
```

```python
arr[3:] # from index 3 to end
```

NUMPY INDEXING AND SLICING OF MATRIX

```python
arr = np.arange(1,31).reshape(6,5)  # 6 rows and 5 columns
arr
```

```python
arr[2] # access third row
```

```python
arr[0,0]  # first row, first column
```

```python
arr[4,3] # fifth row, fourth column
```

```python
arr[0:2, 1:4]  # first two rows, columns 1 to 3
```

```python
arr[3:,3:]
```

```python
arr[:,3]    # all rows, fourth column
```

BOOLEAN INDEXING

```python
ar = np.arange(11,21)
ar
```

```python
bool_index = ar%2==0
bool_index
```

```python
ar = ar[bool_index]
ar
```
