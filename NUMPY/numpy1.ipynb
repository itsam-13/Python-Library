Array generation function


```python
a = np.arange(0, 10)
b = np.arange(0,10,2)
a, b
```

```python
a = np.zeros((3,4)) # 3 rows and 4 columns
a
```

```python
arr = np.ones((3,4)) # 3 rows and 4 columns
arr
```
