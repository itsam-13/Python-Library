<h3><b>ARRAY OPERATIONS</b></h3>

*Arithmatic operations*

```python
a1 = np.array([1,2,3,4,5])
a2 = np.array([1,2,3,4,5])
```

```python
arr2 = a1 + a2 # element-wise addition and have same no of size
```

```python
arr3 = a1 - a2
arr4 = a1 * a2
arr5 = a1 / a2  
arr6 = a1 % a2
arr7 = a1 ** a2
arr8 = a1 // a2
arr2, arr3, arr4, arr5, arr6, arr7, arr8
```

<i>Broadcasting</i>

```python
l = [1,2,3,4,5]
arr = np.array(l)
arr+10
```


<i>Deep and Shallow copy</i>

```python
a = np.arange(1,21)
a
```

```python
slice = a[:5]
slice = slice *10
slice  # slice will take different memory location
```

```python
a # original array remains unchanged -- shallow copy
```

```python
#deep copy 
b = a
b
```

```python
b[0] = 13
b
```

```python
a
```

<i>Matrix Operation</i>

```python
A
```

```python
B
```

```python
A@B  # matrix multiplication
```

```python
np.dot(A,B) # matrix multiplication
```

<h3><b>ðŸ§® Advance Array Manipulation</b></h3>

<i>Stacking Array</i>

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
```

```python
np.vstack((a,b))  # vertical stacking  
```

```python
np.hstack((a,b))  # horizontal stacking
```

```python
np.column_stack((a,b))  # column stacking
```

<i>Splitting Array</i>

```python
c = np.arange(1,17 ).reshape(4,4)
c
```

```python
np.hsplit(c,2)  # horizontal splitting into 2 arrays
```

```python
np.hsplit(c,4)
```

```python
np.vsplit(c,2) # vertical splitting into 2 arrays
```
